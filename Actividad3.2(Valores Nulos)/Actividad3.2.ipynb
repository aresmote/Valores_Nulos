```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('OttawaSelect.csv', encoding='latin1')

df.head(5)
```

```python
#Copia del dataset
df1 = df.copy()
df1
```

```python
#Recuento de los valores nulos del dataframe
nulos = df1.isnull().sum()
nulos = nulos[nulos > 0]
nulos
```

```python
#ELIMINACIÃ“N DE COLUMNA
#Se hizo la eliminaciÃ³n de una columna, ya que todos los registros de esa columna estaban vacÃ­os
df2 = df1.drop(['neighbourhood_group_cleansed'], axis =1)

#Recuento de los valores nulos del dataframe
nulos = df2.isnull().sum()
nulos = nulos[nulos > 0]
nulos
```

**SUSTITUCIÃ“N DE VALORES NULOS DEL DATAFRAME**

```python
#PRIMER MÃ‰TODO DE SUSTITUCIÃ“N 
#SustituciÃ³n de valores nulos por uno string concreto

df2["description"] = df2["description"].fillna("pending description")
df2["neighborhood_overview"] = df2["neighborhood_overview"].fillna("pending description")
df2["host_is_superhost"] = df2["host_is_superhost"].fillna("f")
df2["host_location"] = df2["host_location"].fillna("No location")
df2["neighbourhood"] = df2["neighbourhood"].fillna("Neighborhood highlights")
df2["bathrooms_text"]=df2["bathrooms_text"].fillna("1 bath")
df2["price"]=df2["price"].fillna("$100.00")
df2["host_neighbourhood"]=df2["host_neighbourhood"].fillna("No location")



nulos = df2.isnull().sum()
nulos = nulos[nulos > 0]
nulos
```

```python
#SEGUNDO MÃ‰TODO DE SUSUTITUCIÃ“N
#SusituciÃ³n de valores nulos por valores no nulos hacia adelante
columnas_nulas = ["host_response_time", "host_response_rate", "host_acceptance_rate"]

for columna in columnas_nulas:
    if columna in df2.columns:
        df2[columna].fillna(method="ffill", inplace=True)

#Comprobamos valores nulos
nulos = df2.isnull().sum()
nulos = nulos[nulos > 0]
nulos
```

```python
#TERCER MÃ‰TODO DE SUSUTITUCIÃ“N
#SustituciÃ³n de valores nulos por valores calculados por medio del promedio. 
columnas_nulas = ["review_scores_rating", "reviews_per_month", "review_scores_accuracy", "review_scores_cleanliness", "review_scores_checkin", "review_scores_communication", "review_scores_location", "review_scores_value"]

for columna in columnas_nulas:
    if columna in df2.columns:
        df2[columna].fillna(df2[columna].mean(), inplace=True)

nulos = df2.isnull().sum()
nulos = nulos[nulos > 0]
nulos
```

```python
#CUARTO MÃ‰TODO DE SUSTITUCIÃ“N 
#SustituciÃ³n de valores nulos por un numero especÃ­fico.

df2["bedrooms"]=df2["bedrooms"].fillna(1)
df2["beds"]=df2["beds"].fillna(1)
df2["first_review"]=df2["first_review"].fillna(10)
df2["last_review"]=df2["last_review"].fillna(10)

nulos=df2.isnull().sum()
nulos = nulos[nulos > 0]
nulos
```

```python
#QUINTO MÃ‰TODO DE SUSUTITUCIÃ“N
#SusituciÃ³n de valores nulos por valores no nulos hacia atrÃ¡s

df2["has_availability"]=df2["has_availability"].fillna(method="bfill")

nulos=df2.isnull().sum()
nulos = nulos[nulos > 0]
nulos
```

```python
#IMPORTAR CSV CON LOS VALORES FINALES
df2.to_csv("OttawaSinNulos1.csv")
```
