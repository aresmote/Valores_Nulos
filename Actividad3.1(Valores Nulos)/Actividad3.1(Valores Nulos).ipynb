**MÉTODOS DE SUSTITUCIÓN DE VALORES NULOS**

```python
#REALIZAMOS UNA COPIA DEL DATAFRAME
df2=df.copy()
```

```python
#PRIMER MÉTODO DE SUSUTITUCIÓN PARA LA COLUMNA "salon_ventas" Y "otros_métodos"
#SUSTITUIR VALORES NULOS CON PROMEDIO O MEDIA
df2["salon_ventas"]=df2["salon_ventas"].fillna(round(df["salon_ventas"].mean(),1))
df2["otros_medios"]=df2["otros_medios"].fillna(round(df["otros_medios"].mean(),1))

#VERIFICACIÓN DE LOS VALORES NULOS
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#SEGUNDO MÉTODO DE SUSUTITUCIÓN PARA LAS COLUMNAS "tajetas_debito" y "tarjetas_credito"
#SUSTITUIR VAL9ORES POR UN STRING EN CONCRETO
df2["tarjetas_debito"]=df2["tarjetas_debito"].fillna("XXXXXX")
df2["tarjetas_credito"]=df2["tarjetas_credito"].fillna("XXXXXX")

#Comprobamos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#TERCER MÉTODO DE SUSTITUCIÓN
#SUSTITUIR VALORES NULOS POR UN VALOR NÚMERICO EN CONCRETO 

df2["subtotal_ventas_alimentos_bebidas"]=df2["subtotal_ventas_alimentos_bebidas"].fillna(000.000)

#Comprobamos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#CUARTO MÉTODO DE SUSUTITUCIÓN
#SUSTITUIR VALORES NULOS CON MEDIANA YA QUE ESTOS SOLO CONTENÍAN UN VALOR NULO

columnas_nulas = ["bebidas", "panaderia", "lacteos", "carnes", "verduleria_fruteria", "alimentos_preparados_rotiseria"]

for columna in columnas_nulas:
    if columna in df2.columns:
        df2[columna].fillna(df2[columna].median(), inplace=True)

#Comprobamos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#QUINTO MÉTODO DE SUSUTITUCIÓN
#SUSTITUIR VALORES NULOS POR LOS VALORES NO NULOS HACIA ADELANTE

columnas_nulas = ["almacen", "indumentaria_calzado_textiles_hogar"]

for columna in columnas_nulas:
    if columna in df2.columns:
        df2[columna].fillna(method="ffill", inplace=True)

#Comprobamos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#SEXTO MÉTODO DE SUSUTITUCIÓN
#SUSTITUIR VALORES NULOS POR LOS VALORES NO NULOS HACIA ATRÁS

columnas_nulas = ["electronicos_articulos_hogar", "otros"]

for columna in columnas_nulas:
    if columna in df2.columns:
        df2[columna].fillna(method="bfill", inplace=True)

#Comprobamos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```
